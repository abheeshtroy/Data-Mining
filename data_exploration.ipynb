```python
import pandas as pd

# Load the CSVs
src = pd.read_csv('dataset1_crm.csv')
tgt = pd.read_csv('dataset2_ecommerce.csv')

# Preview
print("CRM dataset preview:")
display(src.head())
print("\nE-commerce dataset preview:")
display(tgt.head())
```

```python
# 1) Column data types
print("CRM data types:")
print(src.dtypes)
print("\nE-commerce data types:")
print(tgt.dtypes)

# 2) Missing values per column
print("\nCRM missing values per column:")
print(src.isnull().sum())
print("\nE-commerce missing values per column:")
print(tgt.isnull().sum())
```

```python
# How many unique values per column?
print("CRM unique counts:")
print(src.nunique())
print("\nE-com unique counts:")
print(tgt.nunique())
```

```python
# Top values for categorical fields
print("CRM status counts:\n", src['status'].value_counts(), "\n")
print("CRM category counts:\n", src['category'].value_counts(), "\n")
print("E-com account_status counts:\n", tgt['account_status'].value_counts(), "\n")
print("E-com tier counts:\n", tgt['tier'].value_counts(), "\n")
print("E-com region counts (top 10):\n", tgt['region'].value_counts().head(10))
```

```python
# Email-domain frequency analysis
src['email_domain'] = src['email_address'].str.split('@').str[1].str.lower()
tgt['email_domain'] = tgt['contact_email'].str.split('@').str[1].str.lower()

print("Top 10 CRM email domains:")
print(src['email_domain'].value_counts().head(10))
print("\nTop 10 E-com email domains:")
print(tgt['email_domain'].value_counts().head(10))

# Check for any weird domains (length > 5 characters, uncommon)
weird_src = src['email_domain'][~src['email_domain'].isin(src['email_domain'].value_counts().head(20).index)]
weird_tgt = tgt['email_domain'][~tgt['email_domain'].isin(tgt['email_domain'].value_counts().head(20).index)]
print("\nUncommon CRM domains:", weird_src.unique())
print("Uncommon E-com domains:", weird_tgt.unique())
```

```python
# Strip non-digits and examine length distributions
src['phone_digits'] = src['phone'].str.replace(r'\D+', '', regex=True)
tgt['phone_digits'] = tgt['telephone'].str.replace(r'\D+', '', regex=True)

print("CRM phone length statistics:")
print(src['phone_digits'].str.len().describe(), "\n")
print("E-com phone length statistics:")
print(tgt['phone_digits'].str.len().describe(), "\n")

# Show counts of each length
print("CRM phone length counts:")
print(src['phone_digits'].str.len().value_counts().sort_index(), "\n")
print("E-com phone length counts:")
print(tgt['phone_digits'].str.len().value_counts().sort_index())
```

```python
# Convert to datetime and explore ranges
src['registration_date'] = pd.to_datetime(src['registration_date'], errors='coerce')
tgt['signup_date']        = pd.to_datetime(tgt['signup_date'], errors='coerce')

print("CRM registration dates range:", src['registration_date'].min(), "→", src['registration_date'].max())
print("E-com signup dates range:",    tgt['signup_date'].min(),    "→", tgt['signup_date'].max())

# Monthly counts
crm_monthly = src['registration_date'].dt.to_period('M').value_counts().sort_index()
ecom_monthly = tgt['signup_date'].dt.to_period('M').value_counts().sort_index()

print("\nCRM registrations by month (first 6):")
print(crm_monthly.head(6))
print("\nE-com signups by month (first 6):")
print(ecom_monthly.head(6))

# Plot trends
import matplotlib.pyplot as plt
plt.figure(figsize=(10,3))
crm_monthly.plot(marker='o', title='CRM registrations per month')
plt.show()

plt.figure(figsize=(10,3))
ecom_monthly.plot(marker='o', title='E-com signups per month')
plt.show()
```

```python
import datetime

# Parse DOB fields
src['birth_date'] = pd.to_datetime(src['birth_date'], errors='coerce')
tgt['date_of_birth'] = pd.to_datetime(tgt['date_of_birth'], errors='coerce')

# Compute age in years (as of today)
today = pd.Timestamp.today()
src['age'] = (today - src['birth_date']).dt.days // 365
tgt['age'] = (today - tgt['date_of_birth']).dt.days // 365

# Quick stats & histogram
print("CRM age stats:\n", src['age'].describe(), "\n")
print("E-com age stats:\n", tgt['age'].describe())

# Plot
import matplotlib.pyplot as plt
plt.figure(figsize=(8,3))
src['age'].hist(bins=20)
plt.title('CRM: Age Distribution')
plt.xlabel('Age (years)')
plt.show()

plt.figure(figsize=(8,3))
tgt['age'].hist(bins=20)
plt.title('E-com: Age Distribution')
plt.xlabel('Age (years)')
plt.show()
```

```python
print("Top 10 CRM job titles:\n", src['job_title'].value_counts().head(10), "\n")
print("Top 10 E-com positions:\n", tgt['position'].value_counts().head(10))
```

```python
from scipy.stats import pearsonr

# Scatter plot
plt.figure(figsize=(5,5))
plt.scatter(src['score'], tgt['rating'], alpha=0.6)
plt.xlabel('CRM score')
plt.ylabel('E-com rating')
plt.title('Score vs. Rating')
plt.show()

# Pearson’s r
r, p = pearsonr(src['score'], tgt['rating'])
print(f"Pearson correlation: r = {r:.2f}, p = {p:.2g}")
```
